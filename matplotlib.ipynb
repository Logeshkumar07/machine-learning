```python
#importing matplotlib library
import matplotlib.pyplot as plt 

#importing numpy
import numpy as np 
x=np.linspace(0,10,100)  #100 evenly spaced values between 0 to 10 
y=np.sin(x)
z=np.cos(x)
print(x)
```

```python
print(y)
```

```python
print(z)
```

```python
#sine wave
plt.plot(x,y)
plt.show()
```

```python
#cose wave
plt.plot(x,z)
plt.show()
```

```python
#adding title, x-axis & y-axis labels

plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel('sine value')
plt.title('sine wave')
plt.show()
```

```python
#parabola
x=np.linspace(-10,10,20)
y=x**2
plt.plot(x,y)
plt.show()
```

```python
#parabola with + design and red color
plt.plot(x, y, 'r+')
plt.show()
```

```python
#for plotting bar graph
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
languages=['english','french','spansih','latin','german']
people=[100,50,150,40,70]  
ax.bar(languages,people) 
plt.xlabel('LANGUAGES')
plt.ylabel('NUMBER OF PEOPLE')
plt.show()
```

```python
#pie chart
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
languages=['english','french','spansih','latin','german']
people=[100,50,150,40,70]  
ax.pie(people, labels=languages,autopct='%1.1f%%')   #%1.1f%% is used to show no. of decimals 
plt.show()
```

```python
#scatter plot
x=np.linspace(0,10,30)
y=np.sin(x)
z=np.cos(x)
fig2=plt.figure()
ax=fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color='g')
ax.scatter(x,z,color='b')
plt.show()
```

```python
#3d scatter plot
fig3=plt.figure()
ax=plt.axes(projection='3d')
z=20*np.random.random(100)
x=np.sin(z)
y=np.cos(z)
ax.scatter(x,y,z,c=z,cmap='Blues')
plt.show()
```
