```python
sns.pairplot(df, hue="species")   #pairplot for species in iris dataset :species incluse setosa versicolor virginica
```

```python
sns.set_theme(style="darkgrid")  #just styling the graph the darkgrid style in background
sns.lineplot(x="sepal_length", y="petal_length", hue="species" ,data=df)
```

```python
sns.barplot(x="sepal_length", y="petal_length", hue="species",capsize=.2 ,data=df)
```
