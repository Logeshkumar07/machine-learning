```python
#importing seaborn 
import seaborn as sns

import matplotlib.pyplot as plt 
import numpy as np
import pandas as pd
```

```python
#total bill vs tip dataset
tips=sns.load_dataset('tips')  #this dataset is taken from cloud
```

```python
tips.head()
```

```python
sns.set_theme()   #setting theme
sns.relplot(data=tips, x='total_bill',y='tip',col='time',hue='smoker',style='smoker',size='size')  #visualizing the data  
```

```python
sns.scatterplot(data=tips,x='total_bill',y='tip',hue='smoker')  #visualizing using scatter plot 
```

```python
iris=sns.load_dataset('iris')       #loaded iris dataset from the cloud
sns.scatterplot(x='sepal_length',y='petal_length',hue='species',data=iris)
```
